```python
class Material:
    def __init__(self, id_material, titulo, categoria):
        self.id_material = id_material
        self.titulo = titulo
        self.categoria = categoria
```

```python
class Libro(Material):
    def __init__(self, id_material, titulo, categoria, autor, genero, estado):
        super().__init__(id_material, titulo, categoria)
        self.autor = autor
        self.genero = genero
        self.estado = estado  # 0: Disponible, 1: Prestado

    def registrar(self):
        estado_libro = "Disponible" if self.estado == 0 else "Prestado"
        print(f" Libro registrado - ID: {self.id_material} - {self.titulo} | Autor: {self.autor} | GÃ©nero: {self.genero} | Estado: {estado_libro}")
```

```python
class Revista(Material):
    def __init__(self, id_material, titulo, categoria, edicion, periodicidad, estado):
        super().__init__(id_material, titulo, categoria)
        self.edicion = edicion
        self.periodicidad = periodicidad
        self.estado = estado

    def registrar(self):
        estado_revista = "Disponible" if self.estado == 0 else "Prestado"
        print(f"ðŸ“° Revista registrada - ID: {self.id_material} - {self.titulo} | EdiciÃ³n: {self.edicion} | Periodicidad: {self.periodicidad} | Estado: {estado_revista}")
```

```python
class MaterialDigital(Material):
    def __init__(self, id_material, titulo, categoria, formato, enlace):
        super().__init__(id_material, titulo, categoria)
        self.formato = formato
        self.enlace = enlace

    def registrar(self):
        print(f" Material Digital registrado - ID: {self.id_material} - {self.titulo} | Formato: {self.formato} | Enlace: {self.enlace}")
```

```python
class Persona:
    def __init__(self, nombre, correo):
        self.nombre = nombre
        self.correo = correo

class Bibliotecario(Persona):
    def __init__(self, nombre, correo, id_trabajador, sucursal):
        super().__init__(nombre, correo)
        self.id_trabajador = id_trabajador
        self.sucursal = sucursal

    def agregar_material(self,catalogo,libro,revista,m_digital,opc):
        self.catalago=catalogo
        self.libro=libro
        self.revista=revista
        self.m_digital=m_digital
        self.opc=opc

        if self.catalago.id != self.sucursal.No_sucursal:
            print(f"el Bibliotecario:{self.nombre}  en la sucursal:{self.sucursal.No_sucursal}")
            print("No puede agregar un material a un catalogo diferente de su sucursal")
        else:
            print(f"el Bibliotecario:{self.nombre}  en la sucursal:{self.sucursal.No_sucursal}")
            print(f"agrego un nuevo material al catalogo:{self.sucursal.No_sucursal}")
            self.catalago.agregar_catalogo_suc(libro,revista,m_digital,opc)

    def confirmar_prestamo(self,prestamo):
        self.prestamo=prestamo

        if self.prestamo.usuario.historial <= 0:
            print(f"usuario {self.prestamo.usuario.nombre} no pudo recivir el material su historial no se lo permite")
        else:
            if self.prestamo.opc == 0:
                print(f"prestamo autorizado  - usuario {self.prestamo.usuario.nombre} material:{self.prestamo.libro.titulo} fecha de prestamo:{self.prestamo.fecha_prestamo} fecha de entrega:{self.prestamo.fecha_devolucion}")
                self.prestamo.libro.estado = 1;
                self.prestamo.usuario.prestamos.append(self.prestamo.libro)
            if self.prestamo.opc == 1:
                print(f"prestamo autorizado  - usuario {self.prestamo.usuario.nombre} material:{self.prestamo.revista.titulo} fecha de prestamo:{self.prestamo.fecha_prestamo} fecha de entrega:{self.prestamo.fecha_devolucion}")
                self.prestamo.revista.estado = 1;
                self.prestamo.usuario.prestamos.append(self.prestamo.revista)
            if self.prestamo.opc == 2:
                print(f"prestamo autorizado  - usuario {self.prestamo.usuario.nombre} material:{self.prestamo.m_digital.titulo} ")
                self.prestamo.m_digital.estado = 1;
                self.prestamo.usuario.prestamos.append(self.prestamo.m_digital)

class Usuario(Persona):
    def __init__(self, nombre, correo, id_usuario, historial):
        super().__init__(nombre, correo)
        self.id_usuario = id_usuario
        self.historial = historial

        self.prestamos = []

    def ver_catalogo(self, catalogo):
        print(f" {self.nombre} (ID: {self.id_usuario}) estÃ¡ viendo el catÃ¡logo de la sucursal {catalogo.id_sucursal}.")
        catalogo.mostrar_materiales()

    def solicitar_prestamo(self, material, bibliotecario):
        if self.historial <= 0:
            print(f" {self.nombre} no puede solicitar prÃ©stamos debido a su historial.")
        else:
            print(f" {self.nombre} ha solicitado el prÃ©stamo de {material.titulo}.")
            self.prestamos.append(material)
            material.estado = 1  # Marcar como prestado

    def mostrar_prestamos(self,nc):
        opc=nc
        print(f" {self.nombre} tiene los siguientes materiales prestados:")
        for material in self.prestamos:
            print(f"- {material.titulo} ({'Libro' if isinstance(material, Libro) else 'Revista' if isinstance(material, Revista) else 'Material Digital'})")

    def devolver_material(self, material, penalizacion):
        if material in self.prestamos:
            retraso = penalizacion.verificar_fecha()
            if retraso:
                print(f" {self.nombre} ha devuelto {material.titulo} tarde. Se reduce su historial.")
                self.historial -= 1
            else:
                print(f" {self.nombre} ha devuelto {material.titulo} a tiempo. Se mejora su historial.")
                self.historial += 1
            self.prestamos.remove(material)
            material.estado = 0  # Marcar como disponible
        else:
            print(f" {self.nombre} no tiene prestado {material.titulo}.")
```

```python
class Catalogo:
    def __init__(self, id_sucursal):
        self.id_sucursal = id_sucursal
        self.materiales = []

    def agregar_material(self, material):
        self.materiales.append(material)

    def mostrar_materiales(self):
        print(f" CatÃ¡logo de la Sucursal {self.id_sucursal}")
        for material in self.materiales:
            if isinstance(material, Libro):
                material.registrar()
            elif isinstance(material, Revista):
                material.registrar()
            elif isinstance(material, MaterialDigital):
                material.registrar()

class Prestamo:
    def __init__(self, usuario, material, fecha_prestamo, fecha_devolucion):
        self.usuario = usuario
        self.material = material
        self.fecha_prestamo = fecha_prestamo
        self.fecha_devolucion = fecha_devolucion

    def registrar_prestamo(self):
        print(f" PrÃ©stamo registrado - Usuario: {self.usuario.nombre} - Material: {self.material.titulo} - Fecha de prÃ©stamo: {self.fecha_prestamo} - Fecha de devoluciÃ³n: {self.fecha_devolucion}")
        self.material.estado = 1  # Marcar como prestado
```

```python
class sucursal:

    def __init__(self,No_sucursal,ciudad,catalogo):
        self.No_sucursal=No_sucursal
        self.ciudad=ciudad
        self.catalogo=catalogo


    def mostar_catalago_suc(self):
        opc=self.catalogo.id
        print(f"Sucursal:{self.No_sucursal} con cede en {self.ciudad}")
        self.catalogo.mostrar_cat(opc)
```

```python
class prestamo:
    def __init__(self,usuario,libro,revista,m_digital,opc,fecha_prestamo,fecha_devolucion):
        self.fecha_prestamo=fecha_prestamo
        self.fecha_devolucion=fecha_devolucion
        self.usuario=usuario
        self.libro=libro
        self.revista=revista
        self.m_digital=m_digital
        self.opc=opc

    def mostrar_prestamo(self):
        if self.opc==0:
            print(f"prestamo de el material {self.libro.id_material} nombre:{self.libro.nombre}")
        if self.opc==1:
            print(f"prestamo de el material {self.revista.id_material} nombre:{self.revista.nombre}")
        if self.opc==2:
            print(f"prestamo de el material {self.m_digital.id_material} nombre:{self.m_digital.nombre}")

```

```python
class catalogo:

    def __init__(self,No_catalago,id):
          self.No_catalago = No_catalago
          self.id=id
          self.lista=[]

    def agregar_catalogo_suc(self,libro,revista,m_digital,opc):
            self.opcion=opc
            self.libro=libro
            self.revista=revista
            self.m_digital=m_digital
            if self.opcion == 0:
                self.lista.append(self.libro)

            if self.opcion == 1:
                self.lista.append(self.revista)

            if self.opcion == 2:
                self.lista.append(self.m_digital)


    def mostrar_cat(self,nc):
        opc=nc
        for libro in self.lista:
            if libro.categoria == 0:
                print(f"id:{libro.id_material} - {libro.titulo} autor:{libro.autor}  genero:{libro.genero}  estado:",end=" ")
                if(libro.estado==0):
                    print("Disponible")
                else:
                    print("Prestado")

        for revista in self.lista:
            if revista.categoria==1:
                print(f"id:{revista.id_material} - {revista.titulo} edicion:{revista.edicion} perdiodo:{revista.periodicidad} estado:",end=" ")
                if(revista.estado==0):print("Disponible")
                else:print("Prestado")

        for m_digital in self.lista:
            if m_digital.categoria==2:
                print(f"id:{m_digital.id_material} - {m_digital.titulo}  categoria:{m_digital.formato}  enlace:{m_digital.enlace}")

    def buscar(self,autor,genero,tp,opc):
        self.autor=autor
        self.genero=genero
        self.tp=tp
        self.opc=opc

        for libro in self.lista:
            if self.opc == 0: #si se quiere buscar por autor
                if(libro.categoria==0 and libro.autor==self.autor): print(f" libro encontrado:{libro.id_material} titulo:{libro.titulo} autor:{libro.autor} genero:{libro.genero}")
            if self.opc == 1: #si se quiere  buscar por genero
                if(libro.categoria==0 and libro.genero==self.genero):print(f" libro encontrado:{libro.id_material} titulo:{libro.titulo} autor:{libro.autor} genero:{libro.genero}")
            if self.opc == 2:#si se quiere bucar por categoria
                if(libro.categoria==0 and libro.categoria==self.tp):print(f" libro encontrado:{libro.id_material} titulo:{libro.titulo} autor:{libro.autor} genero:{libro.genero}")

        for revista in self.lista:
            if self.opc == 2:#si se quiere bucar por categoria
                if(revista.categoria==1 and revista.categoria==self.tp):print(f" Revista encontrada:{revista.id_material} titulo:{revista.titulo} edicion:{revista.edicion}")

        for m_digital in self.lista:
            if self.opc == 2:#si se quiere bucar por categoria
                if(m_digital.categoria==2 and m_digital.categoria==self.tp):print(f" material digital encontrado:{m_digital.id_material} titulo:{m_digital.titulo} enlace:{m_digital.enlace}")
```
