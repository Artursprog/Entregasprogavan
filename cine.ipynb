```python
class Persona:
    def __init__(self, nombre, correo):
        self.nombre = nombre
        self.correo = correo


class Cliente(Persona):
    lista_clientes = []

    def __init__(self, nombre, correo, id_cliente):
        super().__init__(nombre, correo)
        self.id_cliente = id_cliente

    def registrar_cliente(self):
        Cliente.lista_clientes.append(self)
        print(f"Cliente registrado: {self.id_cliente} - Nombre: {self.nombre} - Correo: {self.correo}")

    def reservar_boletos(self, funcion, cantidad, aplicar_descuento, descuento):
        if funcion.capacidad_disponible < cantidad:
            print("No hay suficientes lugares disponibles.")
        else:
            funcion.capacidad_disponible -= cantidad
            print(f" {self.id_cliente} - {self.nombre} ha reservado {cantidad} boletos para la función {funcion.id_funcion}, película: {funcion.pelicula.titulo}")
            if aplicar_descuento:
                print(f"Descuento aplicado: {descuento.descripcion}")


class Trabajador(Persona):
    def __init__(self, nombre, correo, num_empleado, puesto):
        super().__init__(nombre, correo)
        self.num_empleado = num_empleado
        self.puesto = puesto

    def agregar_funciones(self, funcion, id_funcion, pelicula, sala):
        if self.puesto != "gerente":
            print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - No tienes permisos para agregar funciones.")
            return

        funcion.id_funcion = id_funcion
        funcion.pelicula = pelicula
        funcion.sala = sala
        funcion.capacidad_disponible = sala.capacidad

        print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - Función agregada - {funcion.id_funcion}, Película: {funcion.pelicula.titulo}, Sala: {funcion.sala.num_sala}")
        funcion.registrar_funcion()

    def registrar_pelicula(self, pelicula, titulo, duracion, clasificacion, genero):
        if self.puesto != "gerente":
            print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - No puedes registrar películas.")
            return

        pelicula.titulo = titulo
        pelicula.duracion = duracion
        pelicula.clasificacion = clasificacion
        pelicula.genero = genero

        print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - Película registrada: {pelicula.titulo}, Duración: {pelicula.duracion} min, Clasificación: {pelicula.clasificacion}, Género: {pelicula.genero}")
        pelicula.registrar_pelicula()

    def registrar_descuento(self, descuento, descripcion, condiciones):
        if self.puesto != "gerente":
            print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - No puedes registrar descuentos.")
            return

        descuento.descripcion = descripcion
        descuento.condiciones = condiciones

        print(f"Empleado No:{self.num_empleado} - Puesto: {self.puesto} - Descuento registrado: {descuento.descripcion}, Condiciones: {descuento.condiciones}")
        descuento.registrar_descuento()


class Sala:
    def __init__(self, num_sala, tipo, capacidad, disponible):
        self.num_sala = num_sala
        self.tipo = tipo
        self.capacidad = capacidad
        self.disponible = disponible


class Pelicula:
    catalogo_peliculas = []

    def __init__(self, titulo, duracion, clasificacion, genero):
        self.titulo = titulo
        self.duracion = duracion
        self.clasificacion = clasificacion
        self.genero = genero

    def registrar_pelicula(self):
        Pelicula.catalogo_peliculas.append(self)

    @classmethod
    def mostrar_peliculas(cls):
        print("Películas en catálogo:")
        for pelicula in cls.catalogo_peliculas:
            print(f"Título: {pelicula.titulo}")


class Descuento:
    lista_descuentos = []

    def __init__(self, descripcion, condiciones):
        self.descripcion = descripcion
        self.condiciones = condiciones

    def registrar_descuento(self):
        Descuento.lista_descuentos.append(self)

    @classmethod
    def mostrar_descuentos(cls):
        print("Descuentos disponibles:")
        for descuento in cls.lista_descuentos:
            print(f"Descripción: {descuento.descripcion}, Condiciones: {descuento.condiciones}")


class Funcion:
    funciones_disponibles = []

    def __init__(self, id_funcion, pelicula, sala):
        self.id_funcion = id_funcion
        self.pelicula = pelicula
        self.sala = sala
        self.capacidad_disponible = sala.capacidad

    def registrar_funcion(self):
        Funcion.funciones_disponibles.append(self)

    @classmethod
    def mostrar_funciones(cls):
        print("Funciones programadas:")
        for funcion in cls.funciones_disponibles:
            print(f"Función: {funcion.id_funcion}, Película: {funcion.pelicula.titulo}, Sala: {funcion.sala.num_sala}")


# Creación de salas
sala1 = Sala(1, "IMAX", 15, "Disponible")
sala2 = Sala(2, "4DX", 25, "Disponible")
sala3 = Sala(3, "2D", 50, "Disponible")

# Creación de películas
peli1 = Pelicula("Interstellar", 169, "B", "Ciencia Ficción")
peli2 = Pelicula("John Wick 4", 160, "C", "Acción")
peli3 = Pelicula("Soul", 100, "A", "Animación")

# Creación de funciones
funcion1 = Funcion("F001", peli1, sala1)
funcion2 = Funcion("F002", peli2, sala2)
funcion3 = Funcion("F003", peli3, sala3)

# descuetos
descuento1 = Descuento("20% de descuento en martes", "Válido solo los martes en funciones antes de las 6 PM")

#creando esclavos
empleado1 = Trabajador("Carlos", "carlos@email.com", 101, "gerente")
empleado2 = Trabajador("Marta", "marta@email.com", 102, "cajero")


empleado1.registrar_pelicula(peli1, "Interstellar", 169, "B", "Ciencia Ficción")
empleado1.registrar_pelicula(peli2, "John Wick 4", 160, "C", "Acción")

empleado1.agregar_funciones(funcion1, "F001", peli1, sala1)


Funcion.mostrar_funciones()
```
